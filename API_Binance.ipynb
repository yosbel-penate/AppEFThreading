```python
import requests
import json
import pandas as pd
```

```python
url_base='https://www.binance.com/api/v3/klines'
symbol='BTCUSDT'
interval='15m'

symbol_url='?&symbol={}'.format(symbol)
interval_url='&interval={}'.format(interval)
```

```python
url=url_base+symbol_url+interval_url

data = requests.get(url)
dictionary=json.loads(data.text)
```

```python
df=pd.DataFrame.from_dict(dictionary)
df=df.drop(range(6, 12), axis=1)
```

```python
col_names=['time', 'open', 'high', 'low', 'close', 'volume']
df.columns=col_names
```

```python
for col in col_names:
  df[col]=df[col].astype(float)

df['date']=pd.to_datetime(df['time']*100000)
df.head()
```

```python
df.to_csv("2021.csv")
```
